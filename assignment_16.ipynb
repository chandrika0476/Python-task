```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
plt.plot(df['gender'])
```

```python
plt.scatter(x=df['math score'],y=df['writing score'])      #scatter plot graph representation in a graphical way of the matplot 
```

```python
label = list(df['gender'].value_counts().keys())
label
```

```python
value = list(df['gender'].value_counts().values)
value
```

```python
plt.bar(label,value)                                    #if we want to print bar plot
plt.xticks(rotation = 60)
plt.show()
```

```python
label = list(df['lunch'].value_counts().keys())
value = list(df['lunch'].value_counts().values)
plt.bar(label,value)                                    #if we want to print bar plot
plt.xticks(rotation = 60)
plt.show()
```

```python
bin = [0,5,10,15,20]
plt.hist(df['parental level of education'],bins=bin, color = 'black',edgecolor = 'white')                               #histogram isa used to check the distribution of power column
plt.show()
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.relplot(x='gender',y='lunch',data=df,hue='math score',style='parental level of education',col='parental level of education',kind='line')              #this is the relplot graph
```

```python
sns.catplot(x='gender',y='lunch',data=df,kind='violin')      #this is a type of catbolical graph and in catbolical graph thi si violin kind style graph
```

```python
sns.scatterplot(x='gender',y='lunch',data=df)
```

```python
sns.barplot(x='gender',y='lunch',data=df,hue='math score')
plt.title('Performance of students')
plt.ylabel("gender")
plt.xlabel("lunch")
plt.show()
```

```python
sns.catplot(x='gender',y='lunch',data=df,hue='math score')   #this is a catbolical graph
```

```python
sns.catplot(x='gender',y='lunch',data=df,kind='bar',hue='math score')
```

```python
sns.scatterplot(x='gender',y='math score',data=df,hue='lunch')
plt.savefig("Eclipsed")                                              #this is a scatter plot graph
```

```python
sns.boxplot(df["gender"])                     #this is a box plot graph
```
